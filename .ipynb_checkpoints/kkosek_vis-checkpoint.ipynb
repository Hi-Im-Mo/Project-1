```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filepath = 'output_data/RSV_df_cleaned.csv'
rsvlab_df = pd.read_csv(filepath)
```

```python
del rsvlab_df['Unnamed: 0']
```

```python
# Bugged code

# year index 
year_index = ['2016','2017','2018','2019','2020']
# range of months
month_index = np.arange(1, 13)
# summation lists for appending
sum_rsv_detects_list = []
sum_rsv_tests_list = []
# summation counters
sum_rsv_detects = 0
sum_rsv_tests = 0
# blank counter
counter = 0

# Initialize loop in the year index
for x in year_index:
# Run loop through the dataframe
    for y in rsvlab_df:
# Conditional to sort if were in the same month
        if (rsvlab_df.loc[y, rsvlab_df['Year']] == x) and (rsvlab_df.loc[counter,['Year'] == y]):
# Add the values for 'RSV Detections' and 'RSV Tests' into the counter
            sum_rsv_detects += rsvlab_df.loc[counter, 'RSV Detections']
            sum_rsv_tests += rsvlab_df.loc[counter, 'RSV Tests']

        else:
# If month has changed, append all detects and tests to respective lists
            sum_rsv_detects_list.append(sum_rsv_detects)
            sum_rsv_tests_list.append(sum_rsv_tests)
# Reset counters for loop
            sum_rsv_detects = 0
            sum_rsv_tests = 0


            
        
```
